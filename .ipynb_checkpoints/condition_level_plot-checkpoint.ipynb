```python
# Define the color mapping
color_map = {
    "triage": "lightyellow",
    "moderate": "orange",
    "severe": "brown"
}

# Pivot the DataFrame to get the format suitable for stacked bar chart
pivot_df = df.pivot_table(
    index="condition_name", 
    columns="condition_tier", 
    values="condition_level",
    aggfunc='first'
)

# Plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot each condition
for condition in pivot_df.index:
    previous_width = 0
    for tier in ["Primary", "Secondary", "Tertiary"]:
        if pd.isna(pivot_df.loc[condition, tier]):
            color = "white"
        else:
            color = color_map[pivot_df.loc[condition, tier]]
        ax.barh(condition, 1, left=previous_width, color=color, edgecolor='black')
        previous_width += 1

# Add legend
legend_handles = [mpatches.Patch(color=color, label=label) for label, color in color_map.items()]
plt.legend(handles=legend_handles, title="Condition Level")

# Labels and title
plt.xlabel("Health Facility Tier")
plt.ylabel("Condition Name")
plt.title("Condition Levels by Health Facility Tier")
plt.xticks([0.5, 1.5, 2.5], ["Primary", "Secondary", "Tertiary"])

# Display the plot
plt.tight_layout()
plt.show()
```

```python

# Define the color mapping
color_map = {
    "triage": "lightyellow",
    "moderate": "orange",
    "severe": "brown"
}

# Create a function to map the values to colors
def map_colors(val):
    return color_map.get(val, "white")

# Pivot the DataFrame to get the format suitable for heatmap
heatmap_data = df.pivot_table(index="condition_name", columns="condition_tier", values="condition_level", aggfunc=lambda x: ','.join(x))

# Create the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Iterate over the dataframe and add colored rectangles
for i, condition in enumerate(heatmap_data.index):
    for j, tier in enumerate(heatmap_data.columns):
        levels = heatmap_data.loc[condition, tier]
        if pd.notna(levels):
            levels = levels.split(',')
            for k, level in enumerate(levels):
                rect = mpatches.Rectangle(
                    (j, i + k / len(levels)), 1, 1 / len(levels),
                    facecolor=map_colors(level),
                    edgecolor='black'
                )
                ax.add_patch(rect)

# Adjust the ticks and labels
ax.set_xticks(np.arange(len(heatmap_data.columns)) + 0.5)
ax.set_xticklabels(heatmap_data.columns)
ax.set_yticks(np.arange(len(heatmap_data.index)) + 0.5)
ax.set_yticklabels(heatmap_data.index)

# Add the legend
legend_patches = [mpatches.Patch(color=color, label=key) for key, color in color_map.items()]
ax.legend(handles=legend_patches, title="Condition Level")

# Set the labels and title
ax.set_xlabel('Health Facility Tier')
ax.set_ylabel('Condition Name')
ax.set_title('Condition Levels by Health Facility Tier')

# Adjust the plot limits and aspect
ax.set_xlim(0, len(heatmap_data.columns))
ax.set_ylim(0, len(heatmap_data.index))
ax.set_aspect('auto')

plt.gca().invert_yaxis()

plt.show()
```

```python
def print_tree(d, indent='', last='updown'):
    """Recursively prints a tree structure of a nested dictionary."""
    if isinstance(d, dict):
        for i, (key, value) in enumerate(d.items()):
            up = '' if last == 'up' else ''
            down = '' if last == 'down' else ''
            next_last = 'down' if i == len(d) - 1 else 'up'
            print('{}{}'.format(indent, up if i == 0 else down))
            print('{}{}{}'.format(indent, ' ' if i == len(d) - 1 else ' ', key))
            if isinstance(value, dict):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                print_tree(value, next_indent, next_last)
            elif isinstance(value, list):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                for j, item in enumerate(value):
                    if j == len(value) - 1:
                        print('{}{}{}'.format(next_indent, ' ', item))
                    else:
                        print('{}{}{}'.format(next_indent, ' ', item))
                        
def build_tree_string(d, indent='', last='updown'):
    """Recursively builds a tree structure string of a nested dictionary."""
    tree_string = ''
    
    if isinstance(d, dict):
        for i, (key, value) in enumerate(d.items()):
            up = '' if last == 'up' else ''
            down = '' if last == 'down' else ''
            next_last = 'down' if i == len(d) - 1 else 'up'
            tree_string += '{}{}\n'.format(indent, up if i == 0 else down)
            tree_string += '{}{}{}\n'.format(indent, ' ' if i == len(d) - 1 else ' ', key)
            
            if isinstance(value, dict):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                tree_string += build_tree_string(value, next_indent, next_last)
            elif isinstance(value, list):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                for j, item in enumerate(value):
                    if j == len(value) - 1:
                        tree_string += '{}{}{}\n'.format(next_indent, ' ', item)
                    else:
                        tree_string += '{}{}{}\n'.format(next_indent, ' ', item)
    
    return tree_string

def build_tree_string(d, indent='', last='updown', is_top_level=True):
    """Recursively builds a tree structure string of a nested dictionary."""
    tree_string = ''
    
    if isinstance(d, dict):
        for i, (key, value) in enumerate(d.items()):
            up = '' if last == 'up' else ''
            down = '' if last == 'down' else ''
            next_last = 'down' if i == len(d) - 1 else 'up'
            if is_top_level:
                key = f"**{key}**"
            tree_string += '{}{}\n'.format(indent, up if i == 0 else down)
            tree_string += '{}{}{}\n'.format(indent, ' ' if i == len(d) - 1 else ' ', key)
            
            if isinstance(value, dict):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                tree_string += build_tree_string(value, next_indent, next_last, is_top_level=False)
            elif isinstance(value, list):
                next_indent = '{}{}'.format(indent, '    ' if i == len(d) - 1 else '   ')
                for j, item in enumerate(value):
                    if j == len(value) - 1:
                        tree_string += '{}{}{}\n'.format(next_indent, ' ', item)
                    else:
                        tree_string += '{}{}{}\n'.format(next_indent, ' ', item)
    
    return tree_string
```
